```r
# Load necessary libraries
library(GA)

# Define objective functions
func1 <- function(X1, X2, X3, X4) {
  # Cross-sectional area
  2 * X2 * X4 + X3 * (X1 - 2 * X4)
}

func2 <- function(X1, X2, X3, X4) {
  # Static deflection
  60000 / ((X3 * (X1 - 2 * X4)^3) + ((2 * X2 * X4) * ((4 * (X4^2)) + (3 * X1 * (X1 - 2 * X4)))))
}

# Set lower & upper bound;
X1 <- seq(10, 80, by = 0.1)
X2 <- seq(10, 50, by = 0.1)
X3 <- seq(0.9, 5.0, by = 0.1)
X4 <- seq(0.9, 5.0, by = 0.1)


```

```r
# Applied Genetic Algorithm
GA <- ga(type = "real-valued",
         fitness = function(x) -func1(x[1], x[2], x[3], x[4]),
         lower = c(10, 10, 0.9, 0.9), upper = c(80, 50, 5, 5),
         popSize = 50, maxiter = 100, run = 100,
         pcrossover = 0.75,
         pmutation = 0.001)

summary(GA)

plot(GA)
```

```r
# Applied Genetic Algorithm
GA <- ga(type = "real-valued",
         fitness = function(x)-func2(x[1], x[2], x[3], x[4]),
         lower = c(10, 10, 0.9, 0.9), upper = c(80, 50, 5, 5),
         popSize = 50, maxiter = 100, run = 100,
         pcrossover = 0.75,
         pmutation = 0.001)

summary(GA)

plot(GA)
```

```r
a = 0
b = 1
# Function 3
func3 <- function(X1, X2, X3, X4, a, b) {
  (a * func1(X1,X2,X3,X4)) + ( b * func2(X1,X2,X3,X4))
}

# Applied Genetic Algorithm
GA <- ga(type = "real-valued",
         fitness = function(x) -func3(x[1], x[2], x[3], x[4], a, b),
         lower = c(10, 10, 0.9, 0.9, 0, 0), upper = c(80, 50, 5, 5, 1, 1),
         popSize = 50, maxiter = 100, run = 100,
         pcrossover = 0.75,
         pmutation = 0.001)

summary(GA)

plot(GA)
```
